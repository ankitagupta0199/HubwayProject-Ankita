```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
hubway=pd.read_csv('hubway_trips.csv')
```

```python
hubway.isnull().sum()
```

```python
hubway=hubway.drop(['seq_id','status','start_date','end_date','bike_nr','subsc_type','zip_code','birth_date'],axis=1)
```

```python
hubway.isnull().sum()
```

```python
hubway.dropna(inplace=True)
```

```python
#hubway.head(1000)
```

```python
sort_by_start_station = hubway.sort_values('strt_statn',ascending=False)
#print(sort_by_start_station.head(n=1000))
```

```python
plt.figure(figsize=(50,20))
sns.countplot(x='end_statn',hue='gender',data=sort_by_start_station)
```

```python
sort_by_end_station = hubway.sort_values('end_statn',ascending=False).head(1000)
#print(sort_by_end_station.head(n=1000))
```

```python
plt.figure(figsize=(50,20))
sns.countplot(x='end_statn',hue='gender',data=sort_by_end_station)
```

```python
sort_by_duration = hubway.sort_values('duration',ascending=False)
##print(sort_by_duration.head(n=1000))
```

```python
plt.figure(figsize=(50,20))
sns.countplot(x='end_statn',hue='gender',data=sort_by_duration)
```
